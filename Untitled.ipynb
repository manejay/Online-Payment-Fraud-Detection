```python
 import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df=pd.read_csv('onlinefraud.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['type'].unique()
```

```python
df['type']=df['type'].replace('PAYMENT',1)
df['type']=df['type'].replace('TRANSFER',2)
df['type']=df['type'].replace('CASH_OUT',3)
df['type']=df['type'].replace('DEBIT',4)
df['type']=df['type'].replace('CASH_IN',5)
```

```python
df
```

```python
df.columns
```

```python
x=df[['type', 'amount', 'oldbalanceOrg', 'newbalanceOrig','oldbalanceDest', 'newbalanceDest']]
y=df[['isFraud']]
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
	x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import roc_auc_score as ras
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression

model =LogisticRegression()
model.fit(X_train, y_train)

print("Logistic Reggression : ")
	
train_preds = model.predict_proba(X_train)[:, 1]
print('Training Accuracy : ', ras(y_train, train_preds))
	
y_preds = model.predict_proba(X_test)[:, 1]
print('Validation Accuracy : ', ras(y_test, y_preds))
```

# XGBClassifier

```python
from xgboost import XGBClassifier

model =XGBClassifier()
model.fit(X_train, y_train)

print("XGBClassifier : ")
	
train_preds = model.predict_proba(X_train)[:, 1]
print('Training Accuracy : ', ras(y_train, train_preds))
	
y_preds = model.predict_proba(X_test)[:, 1]
print('Validation Accuracy : ', ras(y_test, y_preds))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model =RandomForestClassifier(n_estimators=7,
								criterion='entropy', 
								random_state=7)
model.fit(X_train, y_train)

print("Random Forest Classifier : ")
	
train_preds = model.predict_proba(X_train)[:, 1]
print('Training Accuracy : ', ras(y_train, train_preds))
	
y_preds = model.predict_proba(X_test)[:, 1]
print('Validation Accuracy : ', ras(y_test, y_preds))
```
